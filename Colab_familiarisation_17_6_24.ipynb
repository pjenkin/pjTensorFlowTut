```python
from matplotlib import pyplot as plt

import cartopy
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LatitudeFormatter, LongitudeFormatter
# import from libraries

import matplotlib as mpl    # update matplotlib to avoid error (see SO comment)
mpl.rcParams.update(mpl.rcParamsDefault)  # https://stackoverflow.com/a/69307987

#%matplotlib notebook
#%matplotlib inline
# https://stackoverflow.com/a/64230536

ax = plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax.add_feature(cartopy.feature.RIVERS)
# set formats for feature types

model_netcdf_data['VHM0'].mean(dim='time').plot.pcolormesh(ax=ax, cmap='Blues',
     cbar_kwargs={'label':"$H_s$ [m]", 'location': "bottom", 'extend':"both"})

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
linewidth=2, color='gray', alpha=0., linestyle=':')
gl.top_labels = False
gl.right_labels = False
gl.xformatter = LongitudeFormatter(number_format='.1f')
gl.yformatter = LatitudeFormatter(number_format='.2f')
# set up gridlines of map

```
